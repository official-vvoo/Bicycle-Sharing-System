## Merge Data

```python
# 두 데이터 프레임 행정동명 불일치 데이터(1,000,000개) 전처리  
# bicycle 데이터: **제1동 데이터 X, weather 데이터: **제1동 데이터 O -> 행정동명에서 '제' 제거
for i in range(1,3): 
    weather_df['location'] = weather_df['location'].apply(lambda x: x if ('상일' in x) else x.replace(f'제{i}',f'{i}',1))
    
for i in range(3,9): 
    weather_df['location'] = weather_df['location'].apply(lambda x: x.replace(f'제{i}',f'{i}',1))
    
# bicycle 데이터: **6·7동 , weather 데이터: **6.7동 -> 지도 데이터 기본값인 ·로 통일 
weather_df['location'] = weather_df['location'].apply(lambda x: x.replace('.','·'))
```

```python
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'])
```

```python
# merge
df = pd.merge(df, weather_df, on=['datetime', 'location'], how='left')
```
